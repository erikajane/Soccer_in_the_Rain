```python
import pandas as pd
import sqlite3
```

```python
# Connect to database
conn = sqlite3.connect('''database.sqlite''')

# Create cursor object
cur = conn.cursor()
```

```python
# View Matches dataframe

cur.execute('''SELECT * FROM Matches;''')
Matches_df =pd.DataFrame(cur.fetchall())
Matches_df.columns = [x[0] for x in cur.description]
Matches_df.head()
```

```python
# Create dataframe that shows many goals each team scored during the 2011 season

cur.execute('''SELECT HomeTeam AS TeamName, HomeGoalCount + AwayGoalCount AS TotalGoals
FROM (
SELECT HomeTeam, SUM(FTHG) AS HomeGoalCount
FROM Matches
WHERE Season = '2011'
GROUP BY HomeTeam) AS t1
JOIN (
SELECT AwayTeam, SUM(FTAG) AS AwayGoalCount
FROM Matches
WHERE Season = '2011'
GROUP BY AwayTeam) AS t2
ON t1.HomeTeam = t2.AwayTeam;''')
Total_goal_df =pd.DataFrame(cur.fetchall())
Total_goal_df.columns = [x[0] for x in cur.description]
Total_goal_df.head()
```

```python
# Create a dataframe that shows the number of wins, draws and loses for each team during the 2011 season

cur.execute('''SELECT HomeTeam AS TeamName, (HomeGoalCount + AwayGoalCount) AS TotalGoals, (WinsH+WinsA) AS Wins, (DrawsH+DrawsH) AS Draws, (LosesH+LosesA) AS Loses
FROM (
SELECT HomeTeam, 
SUM(FTHG) AS HomeGoalCount,
SUM(CASE FTR
    WHEN 'H' THEN 1
    WHEN 'D' THEN 0
    WHEN 'A' THEN 0
END) AS WinsH,
SUM(CASE FTR
    WHEN 'H' THEN 0
    WHEN 'D' THEN 1
    WHEN 'A' THEN 0
END) AS DrawsH,
SUM (CASE FTR
    WHEN 'H' THEN 0
    WHEN 'D' THEN 0
    WHEN 'A' THEN 1
END) AS LosesH
FROM Matches
WHERE Season = '2011'
GROUP BY HomeTeam) AS t1
JOIN
(SELECT AwayTeam,
SUM(FTAG) As AwayGoalCount,
SUM(CASE FTR
    WHEN 'H' THEN 0
    WHEN 'D' THEN 0
    WHEN 'A' THEN 1
END) AS WinsA,
SUM(CASE FTR
    WHEN 'H' THEN 0
    WHEN 'D' THEN 1
    WHEN 'A' THEN 0
END) AS DrawsA,
SUM (CASE FTR
    WHEN 'H' THEN 1
    WHEN 'D' THEN 0
    WHEN 'A' THEN 0
END) AS LosesA
FROM Matches
WHERE Season = '2011'
GROUP BY AwayTeam) AS t2
ON
t1.HomeTeam = t2.AwayTeam;''')
Win_lose_df =pd.DataFrame(cur.fetchall())
Win_lose_df.columns = [x[0] for x in cur.description]
Win_lose_df.head()
```
